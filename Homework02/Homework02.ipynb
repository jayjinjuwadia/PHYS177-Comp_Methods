```python
#PROBLEM 1
def Integrate_HW2(f,a,b,N):
    
    h=(b-a)/N
    g=0
    s=0
    for i in range(1,N,2):
        g+=4*f(a+i*h)
    for i in range(2,N,2):
        s+=2*f(a+i*h)
    
    return (1/3)*h*(f(a)+f(b)+s+g)

def test_function(x):
    return 4.0*x**3

Integrate_HW2(test_function,0,1,100)
```

```python
#PROBLEM 2 PART A
from math import pi as pi
from math import sin as sin
from math import cos as cos

def J(m,x):
    def funct_1(q):
        return cos(m*q-x*sin(q))
    z = Integrate_HW2(funct_1,0,pi,1000)
    return (1/pi)*z
```
