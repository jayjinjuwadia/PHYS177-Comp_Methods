```python
 # 1A) Gamma represents the natural death rate of the predators.
 #     Delta represents the growth rate of the predator population.
 #     Alpha is the exponetial growth rate of the prey popultion.
 #     Beta is the death rate of the prey due to predation.
 #     As t -> inf for the prey equation (x) the population goes to infinity. 
 #     This is because the prey have an unlimited food supply meaning that the population 
 #     size is only capped by the rate of predation, which is eliminated in this scenario.
 #     As t -> inf for the predator equation (y) the population goes to zero. This is because the 
 #     predator growth rate is based on the birth rate and the prey growth rate and once this parameter is 
 #     removed the remaining equation is only based on the death rate.

```

```python
#1B)
alpha, beta, delta, gamma = .6, 1.3, 1.0, 1.0

def f(x,y,t):
    return (alpha*x - beta*x*y)

def g(x,y,t):
    return (delta*x*y-gamma*y)
```

```python
print( f(.1,.1,0) )
print( g(.1,.1,0) )
```

```python
##1C
## PART 1

## If you want to use np.arange(...)
import numpy as np

# A reminder of our earlier parameters
alpha, beta, delta, gamma = .6, 1.3, 1.0, 1.0

x0 = 2     # init Caterpie population
y0 = 2     # init Pidgeotto population
dt = 0.05  # time step
N = 100    # number of steps

## A python array: [0,1,2, ..., N]
time = np.arange(N) 



## PART 2

xlist = []
ylist = []

# Initialize x and y
x = x0
y = y0

for t in time:
    xlist.append(x)
    x += dt*f(x,y,0)
    ylist.append(y)
    y += dt*g(x,y,0)



## PART 3

tt = np.array(time)
xx = np.array(xlist)
yy = np.array(ylist)

import matplotlib.pyplot as plt


plt.plot(tt,xx,label='Caterpie')
plt.plot(tt,yy,label='Pidgeotto')
plt.xlabel('time')
plt.ylabel('population')
plt.legend()
plt.show()
```
